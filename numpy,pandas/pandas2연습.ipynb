```python
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
def peak_to_peak_ratio(x):
    return x.max() / x.min()
```

```python
iris
```

```python
iris.groupby(iris.species).size()
iris.groupby(iris.species).count()
iris.groupby(iris.species).mean()
iris.groupby(iris.species).median()
```

```python
iris.groupby(iris.species).agg(peak_to_peak_ratio)
```

```python
iris.groupby(iris.species).describe().T
```

```python
def top3_petal_length(df):
    return df.sort_values(by="petal_length", ascending=False)[:3]

iris.groupby(iris.species).apply(top3_petal_length)
```

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=["소","중","대"]).astype(str)

iris["petal_length_class"] = iris.groupby(iris.species).petal_length.transform(q3cut)
iris
```

```python
iris.groupby(iris.species).apply(lambda x: np.mean(x))
iris.groupby(iris.species).apply(lambda x: x.describe())
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic["연령구간"] = pd.qcut(titanic.age, 3, labels=["젊은", "중년", "노인"])
titanic
```

```python
def pct_alive(x):
    return x.sum() / len(titanic)

titanic.groupby(["sex", "연령구간", "class"]).agg(pct_alive).unstack("class")
titanic
```

```python
titanic.pivot_table("survived",["sex", "class"],aggfunc=pct_alive)
```
